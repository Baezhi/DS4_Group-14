```python
final.CaseRef.nunique()
```

```python
d = pd.read_csv("WDB_DemonicPact.csv")
d.head()
```

```python
dp = d.drop(['DemonicRef', 'DemonicSystemId', 'DemonicID', 'Demonic_Text', 'Createdby', 'Createdate', 'Lastupdatedby', 'Lastupdatedon'], axis = 1)
print(dp)
```

```python
final2 = pd.merge(final, dp)
final2.columns
final2.head()
```

```python
final2.CaseRef.nunique()
```

```python
final2.to_csv (r'Final2', index = False, header=True)
print(final2)
```

```python
merged = final2.drop_duplicates()
print(merged)
```

```python
merged.CaseRef.nunique()
```

```python
merged
```

```python
df1 = merged.groupby('CaseRef').apply(lambda x: x['RitualObject_Type'].unique())
df2 = merged.groupby('CaseRef').apply(lambda x: x['Motif_Type'].unique())
df3 = merged.groupby('CaseRef').apply(lambda x: x['MeetingPlace_Place'].unique())
df4 = merged.groupby('CaseRef').apply(lambda x: x['M_Firstname'].unique())
df5 = merged.groupby('CaseRef').apply(lambda x: x['M_Surname'].unique())
df6 = merged.groupby('CaseRef').apply(lambda x: x['Demonic_Type'].unique())
print(pd.concat([df1,df2,df3,df4,df5,df6], axis=1))
```
